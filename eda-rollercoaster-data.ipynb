# Step5: Ask a Questions about the data
* Trying to answer the questions about the data with the help of plots and statistical methods.

Example Question1: What are the locations with fastest roller coasters (which has minimum of 10 coasters) ?

```python
df.head()
```

```python
# Locations with total number of roller coasters .
df['Location'].value_counts()
```

```python
# Here we are trying to select the loctaions with total number of roller coasters more than 10 and selecting them in decraesing order of their mean speed

df.query('Location != "Other"')\
.groupby('Location')['Speed_mph']\
.agg(['mean','count'])\
.query('count>=10')\
.sort_values('mean', ascending=False)
```

```python
df.query('Location != "Other"')\
.groupby('Location')['Speed_mph']\
.agg(['mean','count'])\
.query('count>=10')\
.sort_values('mean', ascending=False)\
.plot(kind='barh')
```

Plot showing speed and count of roller caosters of the different regions .

```python
ax=df.query('Location != "Other"')\
.groupby('Location')['Speed_mph']\
.agg(['mean','count'])\
.query('count>=10')\
.sort_values('mean')\
['mean'].plot(kind='barh', title ='Average Coaster Speed by location', figsize=(12,6))
ax.set_xlabel('Average Coaster Speed(mph)')
plt.show()
```

Here by this solution of the problem we can conclude that the above plot shows the fastest coasters are in Bush Garden Williamsburg region and we have other locations as well where the average spped can be analysed with above plot. 
Same we will perform for heights in example question 2.

Example Question2: What are the locations with highest roller coasters (which has minimum of 5 coasters) ?
Note- Here taking minimum 5 coasters because heights are not mentioned in many locations of the coassters in our dataset.

```python
df.query('Location != "Other"')\
.groupby('Location')['Height_ft']\
.agg(['mean','count'])\
.query('count>=5')\
.sort_values('mean', ascending=False)
```

```python
df.query('Location != "Other"')\
.groupby('Location')['Height_ft']\
.agg(['mean','count'])\
.query('count>=5')\
.sort_values('mean')\
['mean'].plot(kind='barh', title ='Average Coaster Heights by location', figsize=(12,6))
ax.set_xlabel('Average Coaster Height(feets)')
plt.show()
```

So here we can conclude that Canada's Wonderland has highest roller coasters and other roller coasters heights can be analysed with the plot above.
